Finalmente, graficaré los datos anteriores a manera de histograma con la media de lluvias por mes y año.

```python
LluvMed18 = np.zeros(12)
LluvMed19 = np.zeros(12)
i = 0
for mes in meses:
    LluvMed18[i] = lluv18[mes].mean()
    LluvMed19[i] = lluv19[mes].mean()
    i+=1

plt.title('Precipitación Promedio 2018 y 2019')
plt.plot(meses, LluvMed18, 'r--', alpha =.6, label = '2018')
plt.plot(meses, LluvMed19, 'g--', alpha= .6, label = '2019')
plt.xlabel("Meses")
plt.ylabel('Lluvia Promedio')
plt.legend()
```
