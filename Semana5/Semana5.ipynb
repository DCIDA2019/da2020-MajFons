Gráfica gaussiana, con la expresión $\frac{1}{\sigma \sqrt{2\pi}} exp(-\frac{-(x - \mu)^2}{2 \sigma^2}) $

```python
mu = 70
sigma = 2

x = np.linspace(60, 80, 50000)
Gausiana = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp( - (x - mu) ** 2 / (2 * sigma ** 2))


fig= plt.figure(figsize=(9,7))
plt.hist(np.random.normal(70,2,50000), 50, density = True, alpha = 0.5, color = 'cyan')
plt.plot(x, Gausiana, color = 'blue')

#def fun(m, s):
#    return 
```

Encontré que también podemos usar scipy para graficar la función gausiana y otro tipo de distribuciones muy fácilmente:

```python
x_gauss = np.linspace(60, 80, 50000)
y_gauss = scipy.stats.norm.pdf(x_gauss, mu, sigma)

plt.plot(x_gauss, y_gauss,'-', color = 'blue', markersize = 2, label = 'Gaussiana ($\mu = 70, \sigma = 2$)')

plt.legend()
```

Calculando las CDF:

```python
mu = 70

y_gauss_cum = scipy.stats.norm.cdf(x_gauss, mu, sigma)
plt.plot(x_gauss,y_gauss_cum,'b-', label = 'Normal CDF')
plt.legend()
```

Si y = $\Phi(x) = exp(x)$ y $p(x) = 1$ para $0 <= x <= 1$,la distribución para $x$ es:

```python
x = np.random.uniform(-1,2,500)
p = np.zeros(500)
#print(x[50])
for i in  range(0, len(x)):
    if x[i] >= 0 and x[i] <= 1:
        p[i] = 1
    else:
        p[i] = 0
plt.plot(x, p,'c.')
plt.title("Distribución para x")
plt.xlabel("$x$") 
plt.ylabel("$p(x)$")
```
